```python
Q1. Explain GET and POST methods.
Answer:-
The GET method is used to retrieve information from the server. In Python, you can use the requests.get() method to send a
GET request to a server and retrieve data. For example:
    
import requests

response = requests.get('https://example.com')
print(response.text)

This will send a GET request to https://example.com and print the response content.

The POST method is used to send data to the server, typically to update or create new resources. In Python, you can use the 
requests.post() method to send a POST request to a server. For example:

import requests

data = {'username': 'johndoe', 'password': 'secretpassword'}
response = requests.post('https://example.com/login', data=data)
print(response.status_code)

This will send a POST request to https://example.com/login with the data dictionary containing the username and password 
fields, and print the response status code.

Note that in both cases, the response variable contains the server's response to the request. You can access the response 
headers and content using the response.headers and response.content attributes, respectively.





```

```python
Q3. Why is redirect() used in Flask?
Answer:-The redirect() function is a built-in function in Flask that is used to redirect the user to a different URL. 
It is a useful tool for handling certain scenarios, such as after a form submission or after logging in.

In Flask, you can use the redirect() function to redirect the user to a different URL, which can be either a relative or 
absolute URL. For example, to redirect the user to the homepage of your website, you can use the following code:

python
Copy code
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the homepage!'

@app.route('/about')
def about():
    return 'This is the about page.'

@app.route('/login')
def login():
    # some login logic here
    return redirect(url_for('home'))

if __name__ == '__main__':
    app.run()
In the above example, the login() function redirects the user to the homepage using redirect(url_for('home')) after the user
logs in successfully.

Redirect is a useful tool in Flask because it allows you to handle certain scenarios, such as after a form submission or after
logging in, by redirecting the user to a different URL instead of displaying a static response. This can improve the user 
experience and make your application more dynamic.
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
Answer:- In Flask, templates are HTML files that contain placeholders for dynamic content. Templates enable you to separate the presentation logic from the application logic in your Flask application, making it easier to maintain and modify your code.

The render_template() function is a built-in function in Flask that is used to render templates in your application. It takes the name of the template file as an argument and returns the rendered HTML content. For example:

python
Copy code
from flask import Flask, render_template

app = Flask(__name__)

@app.route('/')
def home():
    return render_template('home.html')

@app.route('/about')
def about():
    return render_template('about.html')

if __name__ == '__main__':
    app.run()
In the above example, the home() and about() functions use the render_template() function to render the home.html and about.html templates, respectively.
```
